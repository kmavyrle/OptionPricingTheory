# Brownian Motion and Quadratic Variation


### Simple Intuition
Brownian motion is essentially a random walk with each step being an infinitesimal increment i.e. a continuous distribution. In random walk, we have stated a discrete number of steps up or down within a time period. However, in Brownian motion, we have partitioned the time range into an infinite amount of partitions, deriving random up or down increments in with each infinitesimally small increase in time.
1. Random Walk represents discrete random movements
2. Brownian motion represents continuous random movements

### Properties of Brownian Motion
**Property 1**
Brownian motion is a Markov process.

##### Simple Intuition: 
What exactly is a Markov process? Put simply, a Markov process is one whereby best predictor of the next outcome given today's outcome is today's outcome. I.e. To rephrase it, the best predictor of tomorrow's price is today's price.
Brownian motion is a Markov process, whereby its conditional distribution of its future state depends only on its current state.
   $$E_{t}[X_{t}] = E[X_{t}|X_{t-1},X_{t-2}\dots X_{0}] = E[X_{t}|X_{t-1}]$$
   $$E[W_{t+\triangle{t}}] = E[W_{t}]+E[W_{\triangle{}t}]$$
   $$=W_{t}+0 = W_{t}$$
   $$E[W_{t+\triangle{t}}] = W_{t}$$



**Property 2**

Brownian motion is nowhere differentiable. 

Intuitively speaking, Brownian motion is an infinitely random process, meaning that no matter how much we keep trying zoom in to a tighter interval of time we will keep seeing a single continuous random squiggly line (i.e the line is not smooth). When the line is not smooth, we are unable to fit a tangent line on it to figure out the slope at any point on the path of a Brownian motion.

From a mathematical standpoint, things get a little more complicated. First, we need to explore the concept of quadratic variation

### Quadratic Variation

##### Simple Intuition
Quadratic variation is akin to second-order differentiation - the rate of change of the rate of change of the dependent variable (nested rate of change). In a smooth function, say $y=x$, its' quadratic variation is zero; meaning $\frac{d^2y}{dx^2} = 0$.

Intuitively speaking, if we have split a particular interval on our x-axis into an infinite amount of partitions such that the size of each partition $\to 0$, the subsequent change in our y-value first-order differential. We can see that the number we obtain would be very small, say 0.00000000....0001, effectively very close to 0 but not 0.

Now in quadratic variation, we square that number. Consequently, it is not difficult to imagine that if we square that number (itself already approaching 0), we would get a corresponding value so small, we can essentially treat it as zero. The aforementioned is what we would obtain for the second-order differential for a smooth function, NOT stochastic processes.


#### Mathematical Intuition

First we start off with the official definition of quadratic variation below:
$$\text{Quadratic Variation} = \sum^n_{i=1}(x_{i}-x_{i-1})^2, \text{ where } P \to 0$$


In the above P refers to the partitions between [0,n]. Looking back at the equation we can tell that if we push P as close as possible to zero, we would likely find that $(x_i-x_{i-1})$ would be very close to zero as well. Consequently $(x_{i}-x{i-1})^2$<span style = color:yellow> would be so negligible that its summation would also be 0.</span>


Mathematically, we have the following:
1. $P \to 0$, where we push the size of each time interval to be as close to 0 as possible which leads to
2. $x_i-x_{i-1} \to 0 \text{ }  \forall \text{ }i$
3. $(x_i-x{i-1})^2 ~ 0\text{ } \forall \text{ }i$


For Brownian motion however, we find that because the Brownian motion process is so noisy that no matter how small we squeeze the time interval to be as close to 0 as possible, the variation is too big for us to assume it to be 0. Consequently, the square of its variation cannot be taken to be 0 as well. This results in us having to apply Taylor's expansion to obtain our second order differential as shown:

$$S_{t} = W_{t}$$
$\text{By Ito's Formula}$
$$dS_{t} = f'(S_{t})dW_{t} + \frac{1}{2}f''(S_{t})(dW_{t})^2$$

Here, the second term is a result of the second-order differential not being equal to zero. $(dW_t)^2$ is our quadratic variation term
